```python
import cv2
image = cv2.imread('image.jpg')
```

```python
image.shape
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
h, w = image.shape[:2]
```

```python
center = (w // 2, h // 2)

# the attriputes are (center_point , angle , scale)
matrix = cv2.getRotationMatrix2D((200,200), -45, 1.0) 

# we got the needed matrix but we didn't pass it to the image yet
# so we use the affine transformation
rotated = cv2.warpAffine(image, matrix, (w, h))
```

```python
cv2_imshow(rotated)
```

```python
cropped_image = rotated[75 : 325, 75 : 325]
cv2_imshow(cropped_image)
```

```python
resize = cv2.resize(cropped_image, (600, 600))
```

```python

cv2_imshow(resize)
```

```python
output = resize.copy()
rectangle = cv2.rectangle(output, (100, 500), 
                          (500, 600), (255, 100, 255), -50)
```

```python
cv2_imshow(output)
```

```python

text = cv2.putText(output, 'Nourhan Waleed', (100, 550), 
                   cv2.FONT_HERSHEY_TRIPLEX, 1.2, (0, 0, 0), 2)
cv2_imshow(output)
```
